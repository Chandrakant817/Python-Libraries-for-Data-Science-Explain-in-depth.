```
df=pd.read_csv("Google-Playstore-32K.csv",nrows=1000)
df.shape
```

```
#taking only Integer and Float values not String
#Taking x-axis as Ratings
#Taking y-axis as Reviews
x=df["Rating"]
y=df["Reviews"]
```

```
plt.scatter(x,y)
```

```
#Applying title
plt.scatter(x,y)
plt.title("Google Playstore Ratings and Reviews")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```
#Applying Figsize
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=100)
plt.title("Google Playstore Ratings and Reviews")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```
#Alpha :-To change the transparency

plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=100,alpha=0.5)
plt.title("Google Playstore Ratings and Reviews")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=100,alpha=0.5,linewidths=10,edgecolors="g")
plt.title("Google Playstore Ratings and Reviews")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```
#explode
#Applying labels
classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors)
plt.show()
```

```
#actoper:-Autopercentage

classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%")
plt.show()
```

```
#shadow

classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True)
plt.show()
```

```
#radius:-To increase or to decrease the radius of pie chart

classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.5)
plt.show()
```

```
#startangle

classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.5,startangle=90)
plt.show()
```

```
#textprops:-Testproperties

classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]
textprops={"fontsize":15}

plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.5,startangle=90,textprops=textprops)
plt.show()
```

```
#startangle

wedgeprops={"linewidth":2,"width":1,"edgecolor":"k"}
classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]
textprops={"fontsize":15}

plt.pie(class1_Students,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct="%0.2f%%",
        shadow=True,
        radius=1,
        startangle=90,
        textprops=textprops,
        pctdistance=0.6,
        counterclock=True,
        wedgeprops=wedgeprops,
        center=(2,3),
        frame=True,
        rotatelabels=True)
plt.show()
```

```
#changing the radius and vaules will also chage the pie chart

wedgeprops={"linewidth":2,"width":1,"edgecolor":"k"}
classes=["Python","R","Machine Learing","Artificial Intelligence","Data Science"]
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0,0,0.2]
colors=["c","b","r","y","g",]
textprops={"fontsize":15}

plt.pie(class1_Students,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct="%0.2f%%",
        shadow=True,
        radius=0.8,
        startangle=270,
        textprops=textprops,
        pctdistance=0.6,
        counterclock=True,
        wedgeprops=wedgeprops,
        center=(2,3),
        frame=True,
        rotatelabels=True)
plt.show()
```
