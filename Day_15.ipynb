```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
tips_df=sns.load_dataset("tips")
tips_df
```

```
tips_df.shape
```

```
sns.distplot(tips_df["size"])
```

```
sns.distplot(tips_df["tip"])
```

```
sns.distplot(tips_df["total_bill"])
```

```
#bins
sns.distplot(tips_df["total_bill"],bins=12)
```

```
#bins
sns.distplot(tips_df["total_bill"],bins=120)
```

```
#KDE
sns.distplot(tips_df["total_bill"],kde=False)
```

```
#rug
sns.distplot(tips_df["total_bill"],rug=True)
```

```
#color
sns.distplot(tips_df["total_bill"],color="y")
```

```
#vertical
sns.distplot(tips_df["total_bill"],vertical=True)
```

```
#norm_hist
sns.distplot(tips_df["total_bill"],norm_hist=True)
```

```
#axlabel we can change the axis label
sns.distplot(tips_df["total_bill"],axlabel="Total Bill")
```

```
#label
sns.distplot(tips_df["total_bill"],label="Total bill",color="m")
plt.title("Histogram of total bill",fontsize=15)
plt.legend()
plt.show()
```
